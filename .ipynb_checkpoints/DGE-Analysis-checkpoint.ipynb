This is a bulk RNA-seq reanalysis of the data provided by Caiado et al. (https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE134506). The input of this file is a count matrix created with the featureCount function of the Subread application.

```r
library(DESeq2)
library(readr)
library(tidyverse)
library(RColorBrewer)
library(pheatmap)
library(DEGreport)
library(tximport)
library(DOSE)
library(pathview)
library(clusterProfiler)
library(ggplot2)
library(ggrepel)
library(gage)
library(gageData)
```

```r
#Read in an annotation file, it was created by the separate script AnnotationFile.ipynb.
annotation <- read.csv("./annotation.csv")
annotation
```

```r
#First we format the count matrix for DESeq2.
data <- read.table("./data/GSE134506_Caiado_RNA-Seq_ProcessedData.txt", sep="	")
data <- data[2:26487, c(1, 7:18)]
colnames(data) <- data[1,]
data <- data[-1,]
rownames(data) <- data[,1]
data <- data[, -1]
data[,] <- apply(data[,], 2, function(x) as.integer(as.character(x)))
data
```

```r
#Delete NA rows (4 pieces).
data <- data[(rowSums(is.na(data)) <= 0),]
data
```

```r
#Create metadata for experiment:
#NT: No treatment. (0.1 uM DMSO)
#Doxo: Doxorubicin 1.8 uM
#Cyta: Cytarabine 6 uM
#Deci: Decitabine 0.1 uM.
treatment <- factor(c(rep("NT", 2), rep("Doxo", 3), rep("Doxo+Cyta", 3), rep("Doxo+Cyta+DAC", 4)))
meta <- treatment %>% data.frame()
rownames(meta) <- colnames(data)
colnames(meta) <- "Treatment"
meta
```

```r
all(colnames(data) %in% rownames(meta))
all(colnames(data) == rownames(meta))
```

```r
## Create DESeq2Dataset object
ddsMat <- DESeqDataSetFromMatrix(countData = data,
                                  colData = meta,
                                  design = ~ Treatment)
```

```r
#To improve the distances/clustering for the PCA and heirarchical clustering visualization methods,
#we need to moderate the variance across the mean by applying the rlog transformation to the normalized counts.
### Transform counts for data visualization
rlogdata <- rlog(ddsMat, blind=TRUE)
rlogdata
```

```r
#Principal component analysis to see whether the samples separate well.
plotPCA(rlogdata, intgroup="Treatment")
```

```r
### Extract the rlog matrix from the object
rlogdata_matrix <- assay(rlogdata)    ## assay() is function from the "SummarizedExperiment" package that was loaded when you loaded DESeq2
### Compute pairwise correlation values
rlogd_correlation <- cor(rlogdata_matrix)    ## cor() is a base R function

head(rlogd_correlation)   ## check the output of cor(), make note of the rownames and colnames
```

```r
#Heatmap representing the calculated correlations between the various samples.
heat.colors <- brewer.pal(6, "Blues")
pheatmap(rlogd_correlation, annotation = meta, color = heat.colors, border_color=NA, fontsize = 10, 
			fontsize_row = 10, height=20)
```

```r
## Run analysis
dds <- DESeq(ddsMat)
```

```r
resultsNames(dds)
```

```r
# Apply fold change shrinkage
res_tableOE <- results(dds, contrast=c("Treatment", "Doxo.Cyta.DAC", "Doxo"), alpha = 0.05) #Overexpressed genes.
res_tableOE <- lfcShrink(dds, coef='Treatment_Doxo.Cyta.DAC_vs_Doxo', res=res_tableOE)
```

```r
results(dds, contrast=c("Treatment", "Doxo.Cyta.DAC", "Doxo"), alpha = 0.05)
```

```r
### Set threshold.
padj.cutoff <- 0.05
```

```r
#Create a table with significantly overexpressed genes.
res_tableOE_tb <- res_tableOE %>%
  data.frame() %>%
  rownames_to_column(var="gene") %>% 
  as_tibble()

sigOE <- res_tableOE_tb %>%
        dplyr::filter(padj < padj.cutoff)
sigOE
```

```r
## Obtain logical vector where TRUE values denote padj values < 0.05 and fold change > 1.5 in either direction

res_tableOE_tb <- res_tableOE_tb %>% 
                  mutate(threshold_OE = padj < 0.05 & abs(log2FoldChange) >= 0.58)
res_tableOE_tb
```

```r
#Create a column to name the top ten genes for visualization.
res_tableOE_tb <- res_tableOE_tb %>% mutate(genelabels = "")
## Sort by padj values 
res_tableOE_tb <- res_tableOE_tb %>% arrange(padj)

res_tableOE_tb$genelabels[1:10] <- res_tableOE_tb$gene[1:10]

res_tableOE_tb
```

```r
ggplot(res_tableOE_tb, aes(x = log2FoldChange, y = -log10(padj))) +
    geom_point(aes(colour = threshold_OE)) +
    geom_text_repel(aes(label = genelabels)) +
    ggtitle("Differential expression under treatment with hypomethilating agent") +
    xlab("log2 fold change") + 
    ylab("-log10 adjusted p-value") +
    theme(legend.position = "none",
          plot.title = element_text(size = rel(1.5), hjust = 0.5),
          axis.title = element_text(size = rel(1.25))) 
```

```r
## Merge the AnnotationHub dataframe with the results 
res_ids <- inner_join(res_tableOE_tb, annotation, by=c("gene"="symbol"))
res_ids
```

```r
#List of all the genes.
all_genes <- res_ids$gene_id %>% as.character()
all_genes
```

```r
#Add ENSEMBL ids to significantly overexpressed genes.
sigOE <- inner_join(sigOE, annotation, by=c("gene"="symbol"))
sig_genes <- sigOE$gene_id %>% as.character()
sig_genes
```

```r
## Run GO enrichment analysis 
ego <- enrichGO(gene = sig_genes, 
                universe = all_genes,
                keyType = "ENSEMBL",
                OrgDb = org.Hs.eg.db, 
                ont = "ALL", 
                pAdjustMethod = "BH", 
                qvalueCutoff = 0.05, 
                readable = TRUE)
```

```r
## Dotplot 
dotplot(ego, showCategory=10)
```

```r
## Enrichmap clusters the 20 most significant (by padj) GO terms to visualize relationships between terms
emapplot(ego, showCategory = 20)
```

```r
#Now we create the objects neccesary for visualization with pathview().
res_entrez <- dplyr::filter(res_ids, is.na(entrezid) == FALSE)

#Removing duplicates. This removes many entries, which is puzzling although it seems to be common practice.
res_entrez <- res_entrez[which(duplicated(res_entrez$entrezid) == F), ]

foldchanges <- res_entrez$log2FoldChange
names(foldchanges) <- res_entrez$entrezid
## Sort fold changes in decreasing order
foldchanges <- sort(foldchanges, decreasing = TRUE)
head(foldchanges)
```

```r
## GSEA using gene sets from KEGG pathways
gseaKEGG <- gseKEGG(geneList = foldchanges, # ordered named vector of fold changes (Entrez IDs are the associated names)
              organism = "hsa", # supported organisms listed below
            #  nPerm = 1000, # default number permutations
              minGSSize = 20, # minimum gene set size (# genes in set) - change to test more sets or recover sets with fewer # genes
              pvalueCutoff = 0.05, # padj cutoff value
              verbose = FALSE)

## Extract the GSEA results
gseaKEGG_results <- gseaKEGG@result
```

```r
gseaKEGG_results
```

```r
#Pathway downloads the KEGG pathway image files and colors them according to the differential expression profile
#of the respective genes.
pathview(gene.data = foldchanges,
              pathway.id = "hsa04110",
              species = "hsa",
              limit = list(gene = 2, # value gives the max/min limit for foldchanges
              cpd = 1))
```

```r
# GSEA using gene sets associated with BP Gene Ontology terms
gseaGO <- gseGO(geneList = foldchanges, 
              OrgDb = org.Hs.eg.db, 
              ont = 'ALL', 
              minGSSize = 20, 
              pvalueCutoff = 0.05,
              verbose = FALSE) 

gseaGO_results <- gseaGO@result
```

```r
gseaGO_results[,-12]
```

```r
#Gage is a different gene set enrichment analysis tool which also found similar results. Gage has the advantage
#of sorting both up- and downregulated genes.
gage(foldchanges, gsets = kegg.gs)
```
