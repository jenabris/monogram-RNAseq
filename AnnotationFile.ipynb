```r
library(AnnotationHub)
library(ensembldb)
```

```r
ah <- AnnotationHub()
human_ens <- query(ah, c("Homo sapiens", "EnsDb"))
human_ens
```

```r
# Extract annotations of interest
human_ens <- human_ens[["AH79689"]]
genes(human_ens, return.type = "data.frame")
```

```r
annotations_ahb <- genes(human_ens, return.type = "data.frame")  %>%
  dplyr::select(gene_id, symbol, entrezid, gene_biotype)
annotations_ahb
```

```r
annotations_ahb$entrezid %>% is.na() %>% summary()
```

```r
annotations_ahb[annotations_ahb$gene_biotype == "protein_coding",]$entrezid %>% is.na() %>% summary()
```

```r
annotations_ahb[annotations_ahb$gene_biotype != "protein_coding",]$entrezid %>% is.na() %>% summary()
```

```r
annotations_ahb$entrezid <- annotations_ahb$entrezid %>%
                            as.character
annotations_ahb
```

It seems that most of the protein coding reads have an EntrezID and most of the non-coding reads do not.
